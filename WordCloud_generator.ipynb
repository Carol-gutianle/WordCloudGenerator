# 中文分词

```python
import re
import jieba
```

```python
# #保留中文内容
# def find_chinese(file):
#     pattern = re.compile(r'[^\u4e00-\u9fa5]')
#     chinese = re.sub(pattern, '', file)
#     return chinese
```

```python
#停用词表设置
stopwords = [i.strip() for i in open('hit_stopwords.txt',encoding='UTF-8').readlines()]
#文字清洗
text = ' '
for str in txt:
    #分词
    cut_list = jieba.lcut(''.join(re.findall('[\u4e00-\u9fa5]',str)))
    for word in cut_list:
        if word not in stopwords:
            text += ' ' + word
text
```

# 生成词云

```python
from wordcloud import WordCloud,ImageColorGenerator
import imageio
```

```python
#导入中文字体 否则乱码
font = r'SimHei.ttf'
#读入背景图片
bg_pic = imageio.imread('bg.png')
# bg_pic = imread('bg.jpg')
```

```python
wcd = WordCloud(mask=bg_pic,background_color="white",repeat=False,max_words=100,font_path=font)
```

```python
wcd.generate(text)
```
